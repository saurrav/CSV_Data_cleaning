# what product are most often sold together

```python
all_data=all_data[all_data['Order ID'].duplicated(keep=False)]

all_data['grouped']=all_data.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))
all_data.head()
```

```python
df=all_data[['Order ID','grouped']].drop_duplicates()
df.head()
```

```python
#to count product numbers
```

```python
from itertools import combinations
from collections import Counter
```

```python
count=Counter()
for row in df['grouped']:
    row_list=row.split(',')
    count.update(Counter(combinations(row_list,2)))
    
for key,value in count.most_common(10):
    print(key,value)
```
